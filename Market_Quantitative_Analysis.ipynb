```python
import pandas as pd
import matplotlib.pyplot as plt
import talib as ta
import pynance as pn

# Load the data
# Replace 'your_stock_data.csv' with your actual CSV file path
df = pd.read_csv('data/AAPL.csv', parse_dates=['Date'])
df.set_index('Date', inplace=True)

# Display first few rows to check
print(df.head())

# Step 3: Calculate technical indicators using TA-Lib
# Simple Moving Average (SMA)
df['SMA_20'] = ta.SMA(df['Close'], timeperiod=20)
df['SMA_50'] = ta.SMA(df['Close'], timeperiod=50)

# Relative Strength Index (RSI)
df['RSI_14'] = ta.RSI(df['Close'], timeperiod=14)

# Moving Average Convergence Divergence (MACD)
df['MACD'], df['MACD_signal'], df['MACD_hist'] = ta.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)

# Step 4: Use PyNance to calculate financial metrics
# Daily Returns
df['Daily_Return'] = df['Close'].pct_change()

# Cumulative Returns
df['Cumulative_Return'] = (1 + df['Daily_Return']).cumprod()

# Step 5: Visualize the data
plt.figure(figsize=(15,10))

# Plot closing price and moving averages
plt.subplot(3,1,1)
plt.plot(df['Close'], label='Close Price')
plt.plot(df['SMA_20'], label='SMA 20')
plt.plot(df['SMA_50'], label='SMA 50')
plt.title('Stock Price with Moving Averages')
plt.legend()

# Plot RSI
plt.subplot(3,1,2)
plt.plot(df['RSI_14'], label='RSI 14', color='orange')
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.title('RSI (Relative Strength Index)')
plt.legend()

# Plot MACD
plt.subplot(3,1,3)
plt.plot(df['MACD'], label='MACD', color='blue')
plt.plot(df['MACD_signal'], label='Signal Line', color='red')
plt.bar(df.index, df['MACD_hist'], label='MACD Histogram', color='gray')
plt.title('MACD (Moving Average Convergence Divergence)')
plt.legend()

plt.tight_layout()
plt.show()

# Step 6: Display financial metrics
print(df[['Daily_Return', 'Cumulative_Return']].tail())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime, timedelta
import requests
from bs4 import BeautifulSoup
import re
import warnings
warnings.filterwarnings('ignore')

# NLP Libraries for Sentiment Analysis
import nltk
from textblob import TextBlob
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer

# Download required NLTK data
try:
    nltk.download('vader_lexicon', quiet=True)
    nltk.download('punkt', quiet=True)
except:
    print("NLTK downloads completed or already available")

class NewsSentimentAnalyzer:
    def __init__(self, symbol, period="6mo"):
        """
        Initialize the news sentiment analyzer
        """
        self.symbol = symbol
        self.period = period
        self.stock_data = None
        self.news_data = None
        self.sentiment_data = None
        self.correlation_results = {}

    def fetch_stock_data(self):
        """
        Fetch stock price data using yfinance
        """
        print(f"Fetching stock data for {self.symbol}...")
        try:
            stock = yf.Ticker(self.symbol)
            self.stock_data = stock.history(period=self.period)

            # Calculate daily returns
            self.stock_data['Daily_Return'] = self.stock_data['Close'].pct_change() * 100
            self.stock_data['Price_Change'] = self.stock_data['Close'].diff()

            print(f"Successfully loaded {len(self.stock_data)} trading days")
            return True

        except Exception as e:
            print(f"Error fetching stock data: {e}")
            return False

    def fetch_news_data(self, sample_size=200):
        """
        Fetch news headlines - Using simulated data for demonstration
        In real scenario, you would use news API like NewsAPI, Alpha Vantage, etc.
        """
        print("Fetching news data...")

        # Simulated news data - Replace this with actual news API calls
        # For demonstration, we'll create realistic sample data
        news_samples = [
            # Positive news
            f"{self.symbol} reports strong quarterly earnings beating estimates",
            f"{self.symbol} announces new innovative product launch",
            f"{self.symbol} stock upgraded by major analysts",
            f"{self.symbol} expands into new markets successfully",
            f"{self.symbol} declares higher than expected dividends",
            f"{self.symbol} partnership drives growth prospects",
            f"{self.symbol} achieves record sales figures",
            f"{self.symbol} receives regulatory approval for new project",

            # Negative news
            f"{self.symbol} misses revenue targets in latest report",
            f"{self.symbol} faces regulatory investigation",
            f"{self.symbol} stock downgraded by analysts",
            f"{self.symbol} announces layoffs and restructuring",
            f"{self.symbol} product recall affects consumer confidence",
            f"{self.symbol} faces supply chain disruptions",
            f"{self.symbol} executive leadership changes announced",
            f"{self.symbol} competition intensifies in key markets",

            # Neutral news
            f"{self.symbol} holds annual shareholder meeting",
            f"{self.symbol} releases quarterly financial statements",
            f"{self.symbol} announces conference call schedule",
            f"{self.symbol} files standard regulatory documents",
            f"{self.symbol} board meeting concludes without major announcements"
        ]

        # Generate realistic news data with dates aligned to stock data
        news_list = []
        if self.stock_data is not None:
            trading_dates = self.stock_data.index

            for date in trading_dates:
                # Random number of news articles per day (0-3)
                num_articles = np.random.randint(0, 4)

                for _ in range(num_articles):
                    headline = np.random.choice(news_samples)
                    sentiment_score = self.analyze_sentiment_textblob(headline)

                    news_list.append({
                        'date': date.date(),
                        'datetime': date,
                        'headline': headline,
                        'sentiment_score': sentiment_score,
                        'sentiment_label': self.get_sentiment_label(sentiment_score)
                    })

        self.news_data = pd.DataFrame(news_list)

        if len(self.news_data) > 0:
            print(f"Generated {len(self.news_data)} news articles across {self.news_data['date'].nunique()} days")
        else:
            print("No news data generated - creating sample data for demonstration")
            # Create fallback sample data
            self.create_sample_news_data()

        return True

    def create_sample_news_data(self):
        """
        Create sample news data for demonstration when no real data is available
        """
        sample_dates = pd.date_range(start=datetime.now() - timedelta(days=180),
                                   end=datetime.now(), freq='D')

        news_list = []
        base_headlines = [
            f"{self.symbol} shows strong growth in emerging markets",
            f"{self.symbol} quarterly results exceed expectations",
            f"{self.symbol} faces challenges in supply chain",
            f"{self.symbol} announces new strategic partnership",
            f"{self.symbol} stock performance remains stable",
            f"{self.symbol} innovation drives market leadership",
            f"{self.symbol} regulatory developments impact operations",
            f"{self.symbol} expands product portfolio successfully",
            f"{self.symbol} market competition intensifies",
            f"{self.symbol} achieves sustainability milestones"
        ]

        for date in sample_dates:
            if np.random.random() > 0.3:  # 70% chance of news each day
                num_articles = np.random.randint(1, 4)
                for _ in range(num_articles):
                    headline = np.random.choice(base_headlines)
                    # Add some variation to headlines
                    variation = ["", " - Market reacts positively", " - Analysts watching closely",
                               " - Investors show confidence", " - Industry impact expected"]
                    full_headline = headline + np.random.choice(variation)

                    sentiment_score = self.analyze_sentiment_textblob(full_headline)

                    news_list.append({
                        'date': date.date(),
                        'datetime': date,
                        'headline': full_headline,
                        'sentiment_score': sentiment_score,
                        'sentiment_label': self.get_sentiment_label(sentiment_score)
                    })

        self.news_data = pd.DataFrame(news_list)
        print(f"Created sample dataset: {len(self.news_data)} articles across {self.news_data['date'].nunique()} days")

    def analyze_sentiment_textblob(self, text):
        """
        Analyze sentiment using TextBlob
        Returns polarity score between -1 (negative) and 1 (positive)
        """
        try:
            blob = TextBlob(text)
            return blob.sentiment.polarity
        except:
            return 0.0

    def analyze_sentiment_vader(self, text):
        """
        Analyze sentiment using VADER (Valence Aware Dictionary and sEntiment Reasoner)
        """
        try:
            sia = SentimentIntensityAnalyzer()
            scores = sia.polarity_scores(text)
            return scores['compound']  # Returns between -1 and 1
        except:
            return 0.0

    def get_sentiment_label(self, score):
        """
        Convert sentiment score to label
        """
        if score > 0.1:
            return "Positive"
        elif score < -0.1:
            return "Negative"
        else:
            return "Neutral"

    def align_and_aggregate_data(self):
        """
        Align news and stock data by date and aggregate daily sentiments
        """
        if self.stock_data is None or self.news_data is None:
            print("Please fetch stock and news data first")
            return False

        print("Aligning and aggregating data...")

        # Aggregate daily sentiment scores
        daily_sentiments = self.news_data.groupby('date').agg({
            'sentiment_score': ['mean', 'count', 'std'],
            'headline': 'count'
        }).round(4)

        # Flatten column names
        daily_sentiments.columns = ['avg_sentiment', 'sentiment_count', 'sentiment_std', 'article_count']

        # Reset index for merging
        daily_sentiments.reset_index(inplace=True)

        # Convert date to datetime for merging with stock data
        daily_sentiments['date'] = pd.to_datetime(daily_sentiments['date'])

        # Prepare stock data for merging
        stock_daily = self.stock_data.reset_index()[['Date', 'Close', 'Daily_Return', 'Price_Change']]
        stock_daily['date'] = stock_daily['Date'].dt.date
        stock_daily['date'] = pd.to_datetime(stock_daily['date'])

        # Merge sentiment data with stock data
        self.sentiment_data = pd.merge(stock_daily, daily_sentiments, on='date', how='left')

        # Fill NaN values for days without news
        self.sentiment_data['avg_sentiment'].fillna(0, inplace=True)
        self.sentiment_data['article_count'].fillna(0, inplace=True)
        self.sentiment_data['sentiment_count'].fillna(0, inplace=True)
        self.sentiment_data['sentiment_std'].fillna(0, inplace=True)

        # Create lagged sentiment variables (news from previous day affecting today's price)
        self.sentiment_data['sentiment_lag1'] = self.sentiment_data['avg_sentiment'].shift(1)
        self.sentiment_data['sentiment_lag2'] = self.sentiment_data['avg_sentiment'].shift(2)

        print(f"Final merged dataset: {len(self.sentiment_data)} records")
        return True

    def calculate_correlations(self):
        """
        Calculate correlation between sentiment scores and stock returns
        """
        if self.sentiment_data is None:
            print("Please align data first")
            return

        print("Calculating correlations...")

        # Filter out days with no news for some analyses
        news_days = self.sentiment_data[self.sentiment_data['article_count'] > 0]

        # Calculate various correlations
        correlations = {
            'Same Day': {
                'pearson': news_days['avg_sentiment'].corr(news_days['Daily_Return']),
                'spearman': news_days['avg_sentiment'].corr(news_days['Daily_Return'], method='spearman')
            },
            'Lag 1 Day': {
                'pearson': news_days['sentiment_lag1'].corr(news_days['Daily_Return']),
                'spearman': news_days['sentiment_lag1'].corr(news_days['Daily_Return'], method='spearman')
            },
            'All Days': {
                'pearson': self.sentiment_data['avg_sentiment'].corr(self.sentiment_data['Daily_Return']),
                'spearman': self.sentiment_data['avg_sentiment'].corr(self.sentiment_data['Daily_Return'], method='spearman')
            }
        }

        self.correlation_results = correlations

        # Print correlation results
        print("\n" + "="*50)
        print("CORRELATION ANALYSIS RESULTS")
        print("="*50)

        for time_frame, methods in correlations.items():
            print(f"\n{time_frame}:")
            for method, value in methods.items():
                print(f"  {method.capitalize()} Correlation: {value:.4f}")

        return correlations

    def create_sentiment_visualizations(self):
        """
        Create comprehensive visualizations for sentiment analysis
        """
        if self.sentiment_data is None:
            print("Please align data first")
            return

        print("Creating visualizations...")

        # Set up the plotting style
        plt.style.use('seaborn-v0_8')
        fig = plt.figure(figsize=(20, 16))

        # 1. Time Series of Stock Price and Sentiment
        ax1 = plt.subplot(3, 2, 1)

        # Plot stock price
        color = 'tab:blue'
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Stock Price ($)', color=color)
        ax1.plot(self.sentiment_data['Date'], self.sentiment_data['Close'], color=color, linewidth=2, label='Stock Price')
        ax1.tick_params(axis='y', labelcolor=color)

        # Plot sentiment on secondary y-axis
        ax2 = ax1.twinx()
        color = 'tab:red'
        ax2.set_ylabel('Sentiment Score', color=color)

        # Only plot days with news
        news_days = self.sentiment_data[self.sentiment_data['article_count'] > 0]
        ax2.scatter(news_days['Date'], news_days['avg_sentiment'], color=color, alpha=0.7, s=50, label='Daily Avg Sentiment')
        ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        ax2.tick_params(axis='y', labelcolor=color)

        ax1.set_title(f'{self.symbol} - Stock Price vs News Sentiment Over Time')
        ax1.legend(loc='upper left')
        ax2.legend(loc='upper right')
        plt.grid(True, alpha=0.3)

        # 2. Daily Returns vs Sentiment Scatter Plot
        ax3 = plt.subplot(3, 2, 2)
        news_days = self.sentiment_data[self.sentiment_data['article_count'] > 0]

        # Color points by sentiment intensity
        colors = np.where(news_days['avg_sentiment'] > 0.1, 'green',
                         np.where(news_days['avg_sentiment'] < -0.1, 'red', 'gray'))

        scatter = ax3.scatter(news_days['avg_sentiment'], news_days['Daily_Return'],
                             c=colors, alpha=0.6, s=60)
        ax3.axhline(y=0, color='black', linestyle='-', alpha=0.3)
        ax3.axvline(x=0, color='black', linestyle='-', alpha=0.3)
        ax3.set_xlabel('Average Daily Sentiment Score')
        ax3.set_ylabel('Daily Return (%)')
        ax3.set_title('Daily Returns vs News Sentiment')

        # Add correlation line
        if len(news_days) > 1:
            z = np.polyfit(news_days['avg_sentiment'], news_days['Daily_Return'], 1)
            p = np.poly1d(z)
            ax3.plot(news_days['avg_sentiment'], p(news_days['avg_sentiment']), "r--", alpha=0.8)

        plt.grid(True, alpha=0.3)

        # 3. Sentiment Distribution
        ax4 = plt.subplot(3, 2, 3)
        sentiment_scores = self.news_data['sentiment_score'].dropna()

        # Create histogram with density curve
        n, bins, patches = ax4.hist(sentiment_scores, bins=20, alpha=0.7, color='skyblue', edgecolor='black', density=True)

        # Add density curve
        from scipy.stats import gaussian_kde
        if len(sentiment_scores) > 1:
            density = gaussian_kde(sentiment_scores)
            xs = np.linspace(sentiment_scores.min(), sentiment_scores.max(), 200)
            ax4.plot(xs, density(xs), 'r-', linewidth=2)

        ax4.axvline(x=0, color='red', linestyle='--', alpha=0.7, label='Neutral')
        ax4.set_xlabel('Sentiment Score')
        ax4.set_ylabel('Density')
        ax4.set_title('Distribution of News Sentiment Scores')
        ax4.legend()
        plt.grid(True, alpha=0.3)

        # 4. Sentiment Labels Pie Chart
        ax5 = plt.subplot(3, 2, 4)
        sentiment_counts = self.news_data['sentiment_label'].value_counts()

        colors = ['green', 'red', 'gray']
        wedges, texts, autotexts = ax5.pie(sentiment_counts.values, labels=sentiment_counts.index,
                                          autopct='%1.1f%%', colors=colors, startangle=90)

        for autotext in autotexts:
            autotext.set_color('white')
            autotext.set_fontweight('bold')

        ax5.set_title('Distribution of Sentiment Labels in News Headlines')

        # 5. News Volume Over Time
        ax6 = plt.subplot(3, 2, 5)
        weekly_news = self.news_data.groupby(pd.Grouper(key='datetime', freq='W'))['headline'].count()

        bars = ax6.bar(weekly_news.index, weekly_news.values, alpha=0.7, color='orange', edgecolor='darkorange')
        ax6.set_xlabel('Week')
        ax6.set_ylabel('Number of News Articles')
        ax6.set_title('Weekly News Article Volume')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3)

        # 6. Correlation Heatmap
        ax7 = plt.subplot(3, 2, 6)

        # Select relevant columns for correlation matrix
        corr_columns = ['Close', 'Daily_Return', 'Price_Change', 'avg_sentiment', 'article_count']
        corr_data = self.sentiment_data[corr_columns].corr()

        mask = np.triu(np.ones_like(corr_data, dtype=bool))
        sns.heatmap(corr_data, mask=mask, annot=True, cmap='coolwarm', center=0,
                   square=True, ax=ax7, fmt='.3f')
        ax7.set_title('Correlation Matrix: Stock Metrics vs Sentiment')

        plt.tight_layout()
        plt.show()

        # Save the figure
        fig.savefig(f'{self.symbol}_sentiment_analysis.png', dpi=300, bbox_inches='tight')
        print(f"Visualizations saved as '{self.symbol}_sentiment_analysis.png'")

    def generate_detailed_report(self):
        """
        Generate a comprehensive analysis report
        """
        if self.sentiment_data is None:
            print("Please complete data analysis first")
            return

        print("\n" + "="*60)
        print(f"NEWS SENTIMENT ANALYSIS REPORT - {self.symbol}")
        print("="*60)

        # Basic statistics
        total_articles = len(self.news_data)
        total_trading_days = len(self.sentiment_data)
        days_with_news = len(self.sentiment_data[self.sentiment_data['article_count'] > 0])

        print(f"\nDATA OVERVIEW:")
        print(f"Total News Articles: {total_articles}")
        print(f"Total Trading Days: {total_trading_days}")
        print(f"Days with News Coverage: {days_with_news} ({days_with_news/total_trading_days*100:.1f}%)")

        # Sentiment statistics
        avg_sentiment = self.news_data['sentiment_score'].mean()
        sentiment_std = self.news_data['sentiment_score'].std()

        print(f"\nSENTIMENT ANALYSIS:")
        print(f"Average Sentiment Score: {avg_sentiment:.4f}")
        print(f"Sentiment Score Std Dev: {sentiment_std:.4f}")

        # Sentiment distribution
        sentiment_labels = self.news_data['sentiment_label'].value_counts()
        print(f"\nSENTIMENT DISTRIBUTION:")
        for label, count in sentiment_labels.items():
            percentage = count / total_articles * 100
            print(f"  {label}: {count} articles ({percentage:.1f}%)")

        # Correlation insights
        print(f"\nCORRELATION INSIGHTS:")
        same_day_corr = self.correlation_results['Same Day']['pearson']
        lag1_corr = self.correlation_results['Lag 1 Day']['pearson']

        print(f"Same-day correlation: {same_day_corr:.4f}")
        print(f"1-day lag correlation: {lag1_corr:.4f}")

        # Interpretation
        print(f"\nINTERPRETATION:")
        if abs(same_day_corr) > 0.2:
            direction = "positive" if same_day_corr > 0 else "negative"
            strength = "strong" if abs(same_day_corr) > 0.5 else "moderate"
            print(f" {strength.capitalize()} {direction} correlation found between news sentiment and stock returns")
        else:
            print(" Weak correlation observed between news sentiment and stock returns")

        if abs(lag1_corr) > abs(same_day_corr):
            print(" Lagged sentiment shows stronger correlation, suggesting delayed market reaction")

        print(f"\nREFERENCES:")
        print(" TextBlob: Simplified Text Processing and Sentiment Analysis")
        print(" NLTK VADER: Valence Aware Dictionary and sEntiment Reasoner")
        print(" Pearson Correlation: Linear correlation coefficient")
        print(" Spearman Correlation: Rank-based correlation coefficient")

        print("="*60)

def main():
    """
    Main function to run the news sentiment analysis
    """
    # Example usage
    symbol = "AAPL"  # Change to any stock symbol

    # Initialize analyzer
    analyzer = NewsSentimentAnalyzer(symbol, period="6mo")

    # Fetch data
    if analyzer.fetch_stock_data():
        analyzer.fetch_news_data()

        # Align and analyze data
        if analyzer.align_and_aggregate_data():
            # Calculate correlations
            analyzer.calculate_correlations()

            # Generate report
            analyzer.generate_detailed_report()

            # Create visualizations
            analyzer.create_sentiment_visualizations()

            print(f"\nNews sentiment analysis completed for {symbol}!")

            # Save results to CSV
            analyzer.sentiment_data.to_csv(f'{symbol}_sentiment_analysis_results.csv', index=False)
            analyzer.news_data.to_csv(f'{symbol}_news_data.csv', index=False)
            print(f"Results saved to CSV files")

if __name__ == "__main__":
    main()
```
